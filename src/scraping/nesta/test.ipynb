```python
# Write code for loading all csv files and creating a single dataframe
import os
import pandas as pd

# Path
folder = "data"

# List of files
files = os.listdir(folder)

# List of dataframes
dfs = []
for file in files:
    if file.endswith(".csv"):
        # skip the first 9 rows
        df = pd.read_csv(os.path.join(folder, file), skiprows=9)
        dfs.append(df)
dfs = (
    pd.concat(dfs, ignore_index=True)
    .rename(
        columns={"Page path and screen class" : "page", "Views": "views"}
    )
    [["page", "views"]]
    .drop_duplicates("page")
    .reset_index(drop=True)
    # drop the row where page = "(not set)"
    .query("page != '(not set)'")
)
```

```python
dfs.sort_values('views', ascending=False).reset_index(drop=True).to_csv('data/nesta_sitemap.csv', index=False)
```
